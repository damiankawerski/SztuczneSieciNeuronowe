```python
M = np.random.random((5,5))
M_norm = (M - M.min()) / (M.max() - M.min() + 1e-12)
M_norm[(M_norm < 0.20)] = 0
print(M_norm)
```

```python
# 

from numpy.lib.stride_tricks import sliding_window_view

img = np.arange(1, 26).reshape(5, 5)
windows = sliding_window_view(img, (3, 3))
sums = windows.sum(axis=(-1, -2))

print("Obraz:\n", img)
print("Suma okien 3x3:\n", sums)

```

```python
# Średnia krocząca 1D

x = np.arange(10)
window = 3

# Suma przesuwna
cumsum = np.cumsum(np.insert(x, 0, 0))
moving_avg = (cumsum[window:] - cumsum[:-window]) / window

print("Średnia krocząca:", moving_avg)
```

```python
import numpy as np

# Przykładowa macierz (np. "obraz")
X = np.arange(25).reshape(5, 5)

# Rozmiar okna kroczącego
window = 3

# Tworzymy okna 3x3 przesuwane po macierzy
windows = np.lib.stride_tricks.sliding_window_view(X, (window, window))

# Liczymy średnią w każdym oknie
moving_avg_2d = windows.mean(axis=(2, 3))

print("Macierz wejściowa:\n", X)
print("\nŚrednia krocząca 2D (okno 3x3):\n", moving_avg_2d)
```
