```python
# 21

X = np.random.random((5,5))
K = np.ones((3,3))

def convolve2d(X, K):
    x_rows, x_cols = X.shape
    k_rows, k_cols = K.shape
    output_rows = x_rows - k_rows + 1
    output_cols = x_cols - k_cols + 1
    output = np.zeros((output_rows, output_cols))
    for i in range(output_rows):
        for j in range(output_cols):
            output[i, j] = np.sum(X[i:i+k_rows, j:j+k_cols] * K)
    return output

print(convolve2d(X, K))
```

```python
import numpy as np
# 26

I = np.random.random((6, 6))

F = np.array([[0, 1, 0],
              [1, -4, 1],
              [0, 1, 0]])

def convolve2d(I, F):
    i_rows, i_cols = I.shape
    f_rows, f_cols = F.shape
    output_rows = i_rows - f_rows + 1
    output_cols = i_cols - f_cols + 1
    output = np.zeros((output_rows, output_cols))
    for i in range(output_rows):
        for j in range(output_cols):
            output[i,j] = np.sum(I[i:i+f_rows, j:j+f_cols] * F)
    
    return output

L = convolve2d(I, F)
print(L)
```

```python
# 28 - averaging with moving window

x = np.random.normal(size=50)

window = 5
cumsum = np.cumsum(np.insert(x, 0, 0))
moving_averages = np.convolve(x, np.ones(window)/window, mode='valid')
print(moving_averages)

```
